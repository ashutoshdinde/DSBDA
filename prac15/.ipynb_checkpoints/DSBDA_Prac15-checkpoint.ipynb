```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("covid_vaccine_statewise.csv")
```

```python
df.describe()
```

```python
first_dose_by_state = df.groupby('State')['First Dose Administered'].sum()
```

```python
print("Number of persons state wise vaccinated for first dose in India:")
print(first_dose_by_state['India'].astype(np.int64))
```

```python
second_dose_by_state = df.groupby('State')['Second Dose Administered'].sum()
```

```python
print("Number of persons state wise vaccinated for first dose in India:")
print(second_dose_by_state['India'].astype(np.int64))
```

```python
males_vaccinated = df['Male(Individuals Vaccinated)'].sum()
print("Number of Males vaccinated:", males_vaccinated)
```
